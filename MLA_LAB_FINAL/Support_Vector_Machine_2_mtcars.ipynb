```python
# importing important libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
# reading csv file in to dataframe
data = pd.read_csv('mtcars.csv')
```

```python
data.head()
```

```python
enc = LabelEncoder()
data.model = enc.fit_transform(data.model)
data.cyl = enc.fit_transform(data.cyl)
data.disp = enc.fit_transform(data.disp)
data.hp = enc.fit_transform(data.hp)
data.mpg = enc.fit_transform(data.mpg)
```

```python
data
```

```python
# spitting the dataset to train and test
X = data.iloc[:,2:4]
y = data.mpg
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = SVC(kernel='rbf',C=2,gamma=0.1)
model.fit(X_train, y_train)
model.get_params()
```

```python
y_predict = model.predict(X_test)
print(accuracy_score(y_test,y_predict))
```
